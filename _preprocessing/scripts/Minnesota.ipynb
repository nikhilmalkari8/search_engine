```python
import pandas as pd
reviews_df = pd.read_json("../Data/rawData/review-Minnesota.json", lines = True)
reviews_df
```

```python
reviews_df['user_id'] = reviews_df['user_id'].astype(object)
```

```python
reviews_df.drop_duplicates(subset=['user_id', 'text'], inplace = True)
```

```python
reviews_df.drop_duplicates(subset = ['user_id', 'gmap_id'], inplace = True)
```

```python
pics_df = reviews_df[~reviews_df['pics'].isna()]
pics_df
```

```python
business_df = pd.read_json("../Data/rawData/meta-Minnesota.json", lines = True, encoding="utf-8")
business_df
```

```python
business_df.drop_duplicates(subset=['gmap_id'], inplace=True)
```

```python
all_df = business_df.merge(pics_df, how = "left", on = ["gmap_id"])
all_df
```

```python
all_df.drop(columns=['resp', 'relative_results', 'state'], inplace=True)
```

```python
all_df.rename(columns = {'name_x': 'business_name', 'name_y': 'user_name', 'gmap_id': 'business_id'}, inplace = True)
```

```python
final_df = all_df[~all_df['user_id'].isna()].copy()
final_df
```

```python
final_df.dtypes
```

```python
final_df.to_json('../Data/rawData/final_Minnesota.json', orient = 'records', indent = 2)
```

```python
import json

f = open('../Data/rawData/final_Minnesota.json')
data = json.load(f)
```

```python
grouped_reviews = {}
for review in data:
    business_id = review['business_id']
    if business_id not in grouped_reviews:
        # initialize an empty list to store reviews of a business
        grouped_reviews[business_id] = {
            "business_id": business_id,
            "business_name": review['business_name'],
            "address": review['address'],
            "description": review['description'],
            "latitude": review['latitude'],
            "longitude": review['longitude'],
            "category": review['category'],
            "avg_rating": review['avg_rating'],
            "num_of_reviews": review['num_of_reviews'],
            "price": review['price'],
            "hours": review['hours'],
            "MISC": review['MISC'],
            "url": review['url'],
            "reviews": []
        }
    # append the review to the corresponding business
    grouped_reviews[business_id]['reviews'].append({
        "user_id": review['user_id'],
        "user_name": review['user_name'],
        "time": review['time'],
        "rating": review['rating'],
        "text": review['text'],
        "pics": review['pics']
    })
```

```python
with open('../Data/Businesses_final/Minnesota_grouped_businesses.json', 'w') as f:
    json.dump(list(grouped_reviews.values()), f, indent = 2)
```
