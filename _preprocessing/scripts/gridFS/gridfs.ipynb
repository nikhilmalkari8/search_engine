```python
from pymongo import MongoClient
import gridfs
import os
from bson.objectid import ObjectId
```

```python
dir_path = '../../images/Businesses'

for filename in os.listdir(dir_path):
    if filename.endswith('.jpeg'):
        new_filename = filename.replace(':', '_')
        os.rename(os.path.join(dir_path, filename), os.path.join(dir_path, new_filename))
```

```python
db = MongoClient().project 

fs = gridfs.GridFS(db) # GridFS instance on the database to use

for imageFile in os.listdir('../../images/Businesses'):
    if not imageFile.startswith('.'):
        # print(imageFile)
        business_id = imageFile.split('.')[0]
        business_id = business_id.replace('_', ':')
        # print(business_id)
        business = db.businesses.find_one({'business_id': business_id})
        # print(business)
        if business:
            with open(f'../../images/Businesses/{imageFile}', 'rb') as f:
                uid = fs.put(f, filename = business_id + '.jpeg')
            db.businesses.update_one(
                {"business_id": business_id},
                {"$set": {"image_id": uid}}
                )
```
